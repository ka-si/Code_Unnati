```python
df.drop_duplicates(inplace = True)
```

```python
x = df.iloc[:, :-1]
x.head(5)
```

```python
y = df["target"]
y.head(3)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.35,random_state=25)
```

```python
from sklearn import linear_model
logr = linear_model.LogisticRegression()
```

```python
logr.fit(x_train,y_train)
```

```python
y_pre = logr.predict(x_test)
```

```python
y_pre
```

```python
from sklearn.metrics import confusion_matrix, precision_score,recall_score,f1_score,accuracy_score,classification_report
```

# Checking Model Preformance 

```python
confusion = confusion_matrix(y_test,y_pre)
confusion
```

```python
f1  = f1_score(y_test,y_pre)
f1
```

```python
pre = precision_score(y_test,y_pre)
pre
```

```python
acc = accuracy_score(y_test,y_pre)
acc
```

```python
print(classification_report(y_test,y_pre))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(x_train, y_train)
knn_predicted = knn.predict(x_test)
knn_conf_matrix = confusion_matrix(y_test, knn_predicted)
knn_acc_score = accuracy_score(y_test, knn_predicted)
print("confussion matrix")
print(knn_conf_matrix)
print("\n")
print("Accuracy of K-NeighborsClassifier:",knn_acc_score*100,'\n')
print(classification_report(y_test,knn_predicted))
```
